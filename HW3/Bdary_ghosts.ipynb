**Problem 1: Gaussian wave with periodic boundary conditions.**

The follwing block contains the complete code. To change the boundary condition one needs to set bc_cond to either 'periodic' or 'outflowing'. The top block of the code contains the necessary parameters. 

**To change the profile:**

*f_avg_arr[N_ghost:-N_ghost] = f_init_avg(x,1)*

In the above line we can set the key to:

1: Gaussian profile
2: Gaussian + Top-hat


**To change the scheme:**

*f_avg_final = ssprk3(f_avg_arr,1)*

In the above line one can set the key to:

1: 2nd ordered centered.
2: 1st order upwind.
3: 3rd order upwind.
4: Monotonicity preserving limit.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import sys
plt.ion()

#The essential parameters for the problem
N_cells = 100  #the number of cells. 
N = N_cells+1  #N+1 is the number of cell edges.
N_ghost = 2
CFL = 0.1  #CFL number
dx = 1.0/N_cells  #grid spacing
sig = 0.05  #sigma for the gaussian in the initial function
v = 1.0
dt = CFL*dx/v
T = 1/v #Length of domain in code units is 1.0
alpha = 4.0  #parameter for defining the MC limited
bc_cond = 'periodic' #implementing periodic boundary

x = np.arange(0,1+dx,dx)  #The cell edges
f_avg_arr = np.zeros(2*N_ghost + N_cells)  #stores N-1 initial cell averaged values
f_avg_final = np.zeros(2*N_ghost + N_cells) #stores N-1 final cell averaged values

#function to define the initial closed form profile (non-averaged)
def f_init_gaussianonly(x):
    return np.exp((-(x-0.3)**2)/(2*(sig**2)))

#function constructing the tophat profile
def f_init_gauss_tophat(x):
    if(x <= 0.6): return f_init_gaussianonly(x)
    else:
        if(x >= 0.7 and x <= 0.9): return 1.0
        else: return 0.0
    
    return init_profile
        
#f_avg_arr = f_init_gauss_tophat(x)

#function to calculate the cell averaged values before beginning time iterations
def f_init_avg(x, f_choice):
    
    f_avg_init = np.zeros(N_cells)
    
    for i in range(N_cells):  #this is the number of cells
        #averaging the function between cell edges
        if (f_choice == 1): f_avg = integrate.quad(f_init_gaussianonly,x[i],x[i+1]) #gaussian profile
        else: f_avg = integrate.quad(f_init_gauss_tophat,x[i],x[i+1])   #gaussian and top hat
        f_avg_init[i] = f_avg[0]/dx  #the values are for the cells centered at (x[i]+x[i+1])/2
 
    return f_avg_init

def minmod(x,y):
    return 0.5*(np.sign(x) + np.sign(y))*np.minimum(np.abs(x), np.abs(y))
        
#function to evaluate the flux discretized flux depeding on the choice of scheme 
def flux_discr(phi, scheme):
    '''phi_jp1 = phi[np.arange(1,N_cells+1,1)%N_cells]	#making necessary adjustments to 
    phi_jm1 = phi[np.arange(-1,N_cells-1,1)%N_cells]	#enforce the periodic B.C
    phi_jm2 = phi[np.arange(-2,N_cells-2,1)%N_cells]'''
    
    phi_jp1 = phi[np.arange(N_ghost + 1, N_ghost + N_cells + 1,1)]	#making necessary adjustments to 
    phi_jm1 = phi[np.arange(N_ghost - 1, N_ghost + N_cells - 1,1)]	#enforce the periodic B.C
    phi_jm2 = phi[np.arange(N_ghost - 2, N_ghost + N_cells - 2,1)]
    
    phi_jp0 = phi[N_ghost:-N_ghost]

    #second order centered
    if(scheme == 1): flux = -v*(phi_jp1 - phi_jm1)/(2*dx)
    #first order upwind
    elif(scheme == 2): flux = -v*(phi_jp0 - phi_jm1)/dx
    #third order upwind biased
    elif(scheme == 3): flux = -v*(phi_jp0/2.0 + phi_jp1/3.0 - phi_jm1 + phi_jm2/6.0)/dx
    #MC limiter scheme
    elif(scheme == 4):
        phi_jp1_MP = phi_jp0 + minmod(phi_jp1 - phi_jp0, alpha*(phi_jp0 - phi_jm1))
        phi_jm1_MP = phi_jm1 + minmod(phi_jp0 - phi_jm1, alpha*(phi_jm1 - phi_jm2))
        
        phi_jp1_3u = (5.0/6.0)*phi_jp0 + (1.0/3.0)*phi_jp1 - (1.0/6.0)*phi_jm1
        phi_jm1_3u = (5.0/6.0)*phi_jm1 + (1.0/3.0)*phi_jp0 - (1.0/6.0)*phi_jm2
        
        phi_jphalf_MC = phi_jp0 + minmod(phi_jp1_3u - phi_jp0, phi_jp1_MP - phi_jp0) 
        phi_jmhalf_MC = phi_jm1 + minmod(phi_jm1_3u - phi_jm1, phi_jm1_MP - phi_jm1)
        
        flux = -v*(phi_jphalf_MC - phi_jmhalf_MC)/dx
        
    else: 
        print('Wrong choice!')
        sys.exit()   
    
        
    return flux

def apply_boundary(phi_wbc, condition):
    
    phi_bc_applied = np.zeros(2*N_ghost + len(phi_wbc))
    
    if(condition == 'periodic'):    

        phi_bc_applied[0:N_ghost] = phi_wbc[-N_ghost:]
        phi_bc_applied[-N_ghost:] = phi_wbc[0:N_ghost]
        
    if(condition == 'outgoing'):
        
        #performing a linear extrapolation at the boundaries
        phi_bc_applied[0:N_ghost] = (phi_wbc[N_ghost+1] - phi_wbc[N_ghost])*(np.arange(N_ghost) - N_ghost) + phi_wbc[N_ghost]
        phi_bc_applied[-N_ghost:] = (phi_wbc[-1] - phi_wbc[-2])*(1+np.arange(N_ghost)) + phi_wbc[-2]
    
    #putting in the central slice
    phi_bc_applied[N_ghost:-N_ghost] = phi_wbc
    
    return phi_bc_applied

#time stepping
def ssprk3(phi,scheme):
    t = 0
    while(t<T):
        phi_np1 = np.zeros(len(phi))
        phi1 = np.zeros(len(phi))
        phi2 = np.zeros(len(phi))
        
        phi1_temp = phi[N_ghost:-N_ghost] + dt*flux_discr(phi,scheme)
        phi1 = apply_boundary(phi1_temp,bc_cond)

        phi2_temp = 0.75*phi[N_ghost:-N_ghost] + 0.25*(phi1[N_ghost:-N_ghost] + dt*flux_discr(phi1,scheme))
        phi2 = apply_boundary(phi2_temp,bc_cond)
        
        phi_np1_temp = (1.0/3.0)*phi[N_ghost:-N_ghost] + (2.0/3.0)*(phi2[N_ghost:-N_ghost] + dt*flux_discr(phi2,scheme))
        phi = apply_boundary(phi_np1_temp,bc_cond)  ##reassigning phi with the phi at next time step
        
        t += dt
        
    return phi
    
#calculating the cell averages with zero paddings on both sides to fill ghost cells
f_avg_arr[N_ghost:-N_ghost] = f_init_avg(x,1)

#obtaining the final time step result
f_avg_final = ssprk3(f_avg_arr,1)

#plotting and comparing the initial and final results after one full period
plt.plot(x[1:]-0.5*dx,f_avg_arr[N_ghost:-N_ghost],label='Initial profile')  #plotting the initial profile
plt.plot(x[1:]-0.5*dx,f_avg_final[N_ghost:-N_ghost],label='After one period')  #plotting the final profile
plt.ylim([-0.5,1.5])
plt.xlim([0,1])
plt.title('N = 100, 2nd order Centered, SSPRK3')
plt.legend()
#plt.title('Second order centered, SSPRK3,CFL = 0.1')
```

**Implementing the outgoing boundary condition**

To implement the outgoing boundary condition, I have used linear extrapolation at the edges using the ghost cells. For this simulation, there are two ghost cells because the stencil needs a j-2 atmost. 

```python
#The essential parameters for the problem
N_cells = 100  #the number of cells. 
N = N_cells+1  #N+1 is the number of cell edges.
N_ghost = 2
CFL = 0.1  #CFL number
dx = 1.0/N_cells  #grid spacing
sig = 0.05  #sigma for the gaussian in the initial function
v = 1.0
dt = CFL*dx/v
T = 1/v #Length of domain in code units is 1.0
alpha = 4.0  #parameter for defining the MC limited
bc_cond = 'outgoing' #implementing periodic boundary

x = np.arange(0,1+dx,dx)  #The cell edges
f_avg_arr = np.zeros(2*N_ghost + N_cells)  #stores N-1 initial cell averaged values
f_avg_final = np.zeros(2*N_ghost + N_cells) #stores N-1 final cell averaged values

#calculating the cell averages with zero paddings on both sides to fill ghost cells
f_avg_arr[N_ghost:-N_ghost] = f_init_avg(x,2)

#obtaining the final time step result
f_avg_final = ssprk3(f_avg_arr,4)

#plotting and comparing the initial and final results after one full period
plt.plot(x[1:]-0.5*dx,f_avg_arr[N_ghost:-N_ghost],label='Initial profile')  #plotting the initial profile
plt.plot(x[1:]-0.5*dx,f_avg_final[N_ghost:-N_ghost],label='After one period')  #plotting the final profile
plt.ylim([-0.5,1.5])
plt.xlim([0,1])
plt.title('N = 100, Monotonicity Preserving Limit, SSPRK3')
plt.legend()
#plt.title('Second order centered, SSPRK3, CFL = 0.1')
```

**Test for outgoing boundary condition**

Apart from the above static plot which shows the final profile is zero throughout (and hence a proof of no residue of the profile in the domain), I also calculated the L2 norm of the final profile. A small enough value ensures that the gaussian and top-hat has been advected out smoothly.

```python
print(np.linalg.norm(f_avg_final[N_ghost:-N_ghost]))
```

This shows that the applied outflowing boundary condition works well.
